## Загрузка данных

Определим функцию для загрузки данных. Это позволит удобно работать локально и загружать данные на любом компьютере.

```python
# определение функции для инициализаци датасета
# функция принимает в качестве аргументов локальный путь, сетевой путь и разделитель для датасета
def fetch_data(local_path, url, sep=',', decimal='.'):
    
    # проверяет существует ли локальный файл csv
    if os.path.exists(local_path):
        
        # если да, то загружает локально
        data = pd.read_csv(local_path, sep=sep, decimal=decimal)
    else:
        
        # если нет, загружает файл из сети с выводом соответствующего уведомления
        print('Local file', local_path, 'does not exist.\nDownloading from', url)
        data = pd.read_csv(url, sep=sep, decimal=decimal)
        
    # функция возвращает объект <DataFrame>
    return data
```

```python
# инициализация объектов <DataFrame> и сохранение их в переменные
# объявление локальных путей к датасетам
market_file_local_path = './market_file.csv'
market_money_local_path = './market_money.csv'
market_time_local_path = './market_time.csv'
money_local_path = './money.csv'

# объявление удаленных путей к датасетам
market_file_url = 'https://code.s3.yandex.net/datasets/market_file.csv'
market_money_url = 'https://code.s3.yandex.net/datasets/market_money.csv'
market_time_url = 'https://code.s3.yandex.net/datasets/market_time.csv'
money_url = 'https://code.s3.yandex.net/datasets/money.csv'

# вызов функции <fetch_data()> для инициализации датасетов
market_file = fetch_data(
    local_path=market_file_local_path,
    url=market_file_url,
)

market_money = fetch_data(
    local_path=market_money_local_path,
    url=market_money_url
)

market_time = fetch_data(
    local_path=market_time_local_path,
    url=market_time_url
)

money = fetch_data(
    local_path=money_local_path,
    url=money_url,
    sep=';',
    decimal = ','
)

# объединим датасеты в словарь для удобства выполнения некоторых действий в циклах
dataset_dict = {
    'market_file' : market_file,
    'market_money' : market_money,
    'market_time' : market_time,
    'money' : money
}
```

Определим функцию для вывода гистограмм распределения количественных признаков.

```python
# функция принимает на вход датасет и список с описательными подписями для оси X
def plot_num_features_hist(df, labels):
    
    # получаем список количественных признаков
    num_columns = df.select_dtypes(include='number').columns.tolist()
    
    # задаем область рисования
    fig, axs = plt.subplots(
        
        # рисуем в одну колонку
        ncols=1,
        
        # количество рядов соответствует количеству признаков
        nrows=len(num_columns),
        
        # выделяем по 5 дюймов на одну гистограмму
        figsize=(10, len(num_columns) * 5)
    )
    
    # вложенная функция рисует гистограмму однократно для одного признака
    def plot_hist_once(column_name, label, ax):
        sns.histplot(
            data=df[column_name],
            color=BAR_COLOR,
            ax=ax
        )
        
        # устанавливаем подписи оси Y графика
        ax.set_ylabel('Количество значений')
        
        # этот код для автопереноса подписи оси X
        #wrapped_ylabel = "\n".join(textwrap.wrap(label, width=60))

        # устанавливаем подписи оси X графика
        ax.set_xlabel(label)

        # устанавливаем в качестве названия графика название столбца
        title = f'Feature name: {column_name}'
        ax.set_title(title, fontsize=14, fontweight='bold')
        pass

    # если в датасете один признак для вывода, рисуем гистограмму 1 раз
    if len(num_columns) == 1:
        plot_hist_once(column_name=num_columns[0], label=labels[0], ax=axs)
        pass
        
    # если признаков больше 1, рисуем гистограмму в цикле для каждого прзнака
    else:
        
        # задаем эстетический пробел между графиками
        fig.subplots_adjust(hspace=0.4) 
    
        # в цикле выводим гистограммы для каждого признака
        for i, ax in enumerate(axs):
            plot_hist_once(column_name=num_columns[i], label=labels[i], ax=ax)
            pass
    
    # убираем лишнее
    sns.despine(left=True)
    plt.show()
    pass
```

## Использование пайплайнов

### Обоснование выбора метрики оценки качества моделей

Исходя из задачи исследования, в интересах бизнеса необходимо наибоее полно определить клиентов, у которых высока вероятность снижения покупательской активности, чтобы персонализирвать для них предложения. Т. е. модель не должна определять клиентов с целевой меткой `Снизилась (1)` как клиентов с целевой меткой `Осталась прежней (0)`. При этом персонализация предложений для части клиентов с низкой вероятностю "оттока" не несет существенных издержек. Таким образом, необходимо минимизировать **Ошибку 2-го рода**, т.е. ложноотрицательные ответы модели. Подходящая метрика - **RECALL**.

### Обучение моделей с использованием пайплайнов

Наименование моделей и перечень гиперпараметров.

|**N п/п**| **Модель с фиксированными гиперпараметрами** | **Гиперпараметры для перебора** | **Значения гиперпараметров для перебора** |
| :---: | :--- | :--- | :---: |
|1.|<pre>DecisionTreeClassifier(<br>    random_state=RANDOM_STATE<br>)</pre>|<pre>min_samples_split</pre>|от 2 до 4|
|||<pre>min_samples_leaf</pre>|от 1 до 2|
|||<pre>max_depth</pre>|от 2 до 3|
|||<pre>max_features</pre>|от 2 до 3|
|2.|<pre>KNeighborsClassifier()</pre>|<pre>n_neighbors</pre>|от 2 до 4|
|3.|<pre>LogisticRegression(<br>    class_weight='balanced',<br>    random_state=RANDOM_STATE,<br>    solver='liblinear',<br>    penalty='l1'<br>)</pre>|<pre>C</pre>|от 3 до 8|
|4.|<pre>SVC(<br>    random_state=RANDOM_STATE<br>)</pre>|<pre>degree</pre>|от 1 до 4|
|||kernel|<pre>'poly'<br>'rbf'</pre>|
|||С|от 1 до 3|

- Для каждого пункта таблицы (каждой модели) используем инструмент отборра признаков `SelectKBest(score_func=f_classif)` со значениями `k` (количество признаков) от 15 до 20 включительно (20 это количество признаков после кодирования).
- Для каждого пункта таблицы (каждой модели) используем инструмент для маштабирования `StandardScaler()`, `MinMaxScaler()` и без масштабирования.
- Для `DecisionTreeClassifier()`, `KNeighborsClassifier()` и `SVC()` при подготовке признаков дополнительно  используем `OrdinalEncoder()`.

Разделим датасет `market_full` на входные признаки и целевой.

```python
X = market_full.drop('покупательская_активность', axis=1)
y = market_full['покупательская_активность']
```

```python
# делим выборку на тренировочную и тестовую
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    # целевая переменнв несбалансирована, указываем стратификацию
    stratify=y
)

# проверяем размерности
print('Размерность X_train', X_train.shape)
print('Размерность X_test', X_test.shape)
print('Размерность y_train', y_train.shape)
print('Размерность y_test', y_test.shape)
```

```python
# инициализируем OneHotEncoder()
ohe_ecnoder = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)

# инициализируем OrdinalEncoder()
ord_ecnoder = OrdinalEncoder(categories=[[False, True]], handle_unknown='use_encoded_value', unknown_value=np.nan)
```

```python
# инициализируем ColumnTransformer() без OrdinalEncoder
data_preprocessor_without_ord = ColumnTransformer(
    [
        # передаем OneHotEncoder()
        ('ohe', ohe_ecnoder, X_train.select_dtypes(exclude='number').columns.to_list()),
        # классы для масштабирования укажем позже в grid_params
        # (на мой взгляд так более лаконично, потому, 
        # что инициализированный здесь скелер никак не будет использован)
        ('num', None, X_train.select_dtypes(include='number').columns.to_list())
    ], 
    remainder='passthrough'
)
```

```python
# инициализируем списки с названиями признаков (категориальных и количественных)
ohe_columns = X_train.select_dtypes(exclude='number').columns.to_list()
ohe_columns.remove('акционные_покупки')
ord_columns = ['акционные_покупки']

# инициализируем ColumnTransformer() c OrdinalEncoder
data_preprocessor_with_ord = ColumnTransformer(
    [
        # передаем OneHotEncoder()
        ('ohe', ohe_ecnoder, ohe_columns),
        ('ord', ord_ecnoder, ord_columns),
        # классы для масштабирования укажем позже в grid_params
        # (на мой взгляд так более лаконично, потому, 
        # что инициализированный здесь скелер никак не будет использован)
        ('num', None, X_train.select_dtypes(include='number').columns.to_list())
    ], 
    remainder='passthrough'
)
```

```python
# инициализируем Pipeline
pipeline = Pipeline(
    [
        # та же логика, что и в случае с ColumnTransformer()
        ('preprocessor', None),
        # инициализируем селектор для отбора признаков
        ('selector', SelectKBest(score_func=f_classif)),
        # та же логика, что и в случае с ColumnTransformer()
        ('classifier', None)
    ]
)
```

```python
# инициализируем список параметров (количество отбираемых признаков для перебора)
selector_k_range = range(15, 21)
```

```python
# инициализируем список скелеров
scalers = [StandardScaler(), MinMaxScaler(), 'passthrough']
```

```python
# инициализируем список column_transformers
column_transformers = [data_preprocessor_with_ord, data_preprocessor_without_ord]
```

```python
# инициализируем словари с гиперпараметрами
param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'classifier' : [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'classifier__min_samples_split' : range(2, 4),
        'classifier__min_samples_leaf' : range(1, 3),
        'classifier__max_depth' : range(2, 4),
        'classifier__max_features' : range(2, 4),
        'preprocessor' : column_transformers,
        'preprocessor__num' : scalers,
        'selector__k' : selector_k_range
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'classifier' : [KNeighborsClassifier()],
        'classifier__n_neighbors' : range(2, 5),
        'preprocessor' : column_transformers,
        'preprocessor__num' : scalers,
        'selector__k' : selector_k_range
    },

    # словарь для модели LogisticRegression()
    {
        'classifier' : [
            LogisticRegression(
                # целевая переменная несбалансирована
                class_weight='balanced',
                random_state=RANDOM_STATE,
                solver='liblinear',
                penalty='l1'
            )
        ],
        'classifier__C' : range(2, 7),
        'preprocessor' : [data_preprocessor_without_ord],
        'preprocessor__num' : scalers,
        'selector__k' : selector_k_range
    },
    
    # словарь для модели SVC()
    {
        'classifier': [SVC(probability=True, random_state=RANDOM_STATE)],
        'classifier__kernel': ['rbf', 'poly'],
        'classifier__degree': range(1, 4),
        'classifier__C': range(1, 4),
        'preprocessor' : column_transformers,
        'preprocessor__num': scalers,
        'selector__k' : selector_k_range
    }
]
```

```python
# инициализация объекта GridSearchCV
grid = GridSearchCV(
    pipeline,
    param_grid=param_grid,
    # используем кросс-валидацию на 5 фолдов
    cv=5,
    # в качестве метрики указываем recall
    scoring='recall',
    n_jobs=-1
)
```

```python
# обучение модели на тренировочных данных
grid.fit(X_train, y_train);
```

### Изменение пределов отнесения к классам для повышения метрики

Можно попробовать увеличить метрику **RECALL** путем снижения порога, так как для нас важна полнота. Построим матрицу ошибок модели.

```python
# построение матрицы ошибок
cm = confusion_matrix(y_test, y_pred)
ax = sns.heatmap(cm, annot=True, fmt='d', cmap='flare_r')
ax.set_title('Матрица ошибок при пределе отнесения к классам 0.5\n', fontsize=14, fontweight='bold')
plt.ylabel('True label')
plt.xlabel('Predicted')
plt.show()
```

Для минимизации ошибки второго рода попробуем снизить порог отнесения покупательской активности к классу `Снизилась (1)`. Будем понишать предел вероятности отнесения к классу 1 постепенно, до тех пор, пока метрика **RECALL** не станет равна единице. Для этого определим функцию `get_max_recall_conditions()`. Она позволит рассчитать максимальное значение предела, при котором метрика **RECALL** будет наилучшая.

```python
# функция принимает на вход истиные значения целевой переменной и предсказанные вероятности
def get_max_recall_conditions(y_true, y_predict_proba):
    
    # запускаем цикл по значениям предела с уменьшением от 0.5 до 0.0 включительно
    for threshold in reversed(np.linspace(0.0, 0.5, num=500, endpoint=True)):
        
        # на каждом шаге получаем прогнозные метки исходя из
        # предсказанных вероятностей и текущего значения предела
        current_y_predictions = (y_predict_proba[:, 1] > threshold).astype(int)
        
        # если на данном шаге метрика recall равна единице, завершаем цикл и возвращаем из функции
        # текущее значение предела и целевые метки классов для данного значение предела
        if recall_score(y_true, current_y_predictions) == 1:
            return threshold, current_y_predictions
```

Сначала посмотрим как рапределены вероятности отнесения наблюдений к классам на тестовой выборке. Выведем гистограмму частотности.

```python
# получаем вероятности
y_proba = pipeline.predict_proba(X_test)

# выводим гистограмму
ax = sns.histplot(y_proba[:, 1], bins=40, color=BAR_COLOR)

# устанавливаем название и подписи осей графика
ax.set_title('Распределение вероятностей отнесения наблюдений к классам', fontsize=14, fontweight='bold')
ax.set_ylabel('Количество наблюдений')
ax.set_xlabel('Вероятность отнесения к классу 1')
    
# убираем лишнее
ax.grid(False)
sns.despine(left=True)
plt.show()
```

По гистограмме видно, что крайних значений сильно больше, чем занчений в районе 0.5. Это значит, что модель уверенно присваивает метки. Придется очень сильно двигать порог для достижения максимального показателя метрики **RECALL**. При этом мы получим большую ошибку 1-го рода.

```python
threshold, y_pred_threshold = get_max_recall_conditions(y_test, y_proba)
print('Предел отнесения наблюдения к классу 1 для максимуму RECALL:', threshold)
```

Посмотрим на матрицу ощибок при таком пределе.

```python
# построение матрицы ошибок
cm = confusion_matrix(y_test, y_pred_threshold)
ax = sns.heatmap(cm, annot=True, fmt='d', cmap='flare_r')
ax.set_title(
    f'Матрица ошибок при пределе отнесения к классам {round(threshold, 3)}\n',
    fontsize=14,
    fontweight='bold'
)
plt.ylabel('True label')
plt.xlabel('Predicted')
plt.show()
```

Ожидаемо мы получили огромную ошибку 1-го рода (**195**). Оставим пределе на дефолтном значении (**0.5**).

### Оценка важности признаков для лучшей модели

Получаем наборы признаков, отобранных и подготовленных для обучения лучшей модели. Определим функцию для подготовки признаков.

```python
# функция принимает на вход объект набор признаков, объект ColumnTransformer (fitted)
# и селектор для отбора признаков (fitted)
def prepare_features(X, preprocessor, selector):
    # инициализируем DataFrame с тренировочным набором
    X_transformed_selected = pd.DataFrame(
    
        # к шагам пайплайна обращаемся по ключам
        # из Pipeline получаем ColumnTransformer, кодируем и масштабируем признаки
        data=preprocessor.transform(X),
    
        # из ColumnTransformer получаем новые названия признаков
        columns=pipeline['preprocessor'].get_feature_names_out()
    
        # применяем маску селектора
    ).loc[:, selector.get_support()]
    # функция возвращаем подготовленные для модели набор данных
    return X_transformed_selected
```

```python
# вызываем ранее определенную функцию для подготовки данных
X_train_transformed_selected = prepare_features(
    X=X_train,
    preprocessor=pipeline['preprocessor'],
    selector=pipeline['selector']
)
```

```python
# смотрим на общую информацию
X_train_transformed_selected.info()
```

После кодирования (OneHotEncoder()) и отбора признаков (SelectKBest()) инмеется 18 признаков для обучения модели. Из них 7 ohe и 11 количественных (num). Визуализируем вклад признаков в каждое предсказание модели методом **SHAP**. Определим функции для вывода графиков.

```python
# функция принимает на вход модель и подготовленный для нее набор признаков
def plot_shap_beeswarm(model, X_transformed_selected):
    # используем библиотеку shap
    explainer = shap.LinearExplainer(model, X_transformed_selected)
    shap_values = explainer(X_transformed_selected)
    shap.plots.beeswarm(
        shap_values,
        
        # передаем размерность в max_display
        max_display=len(X_transformed_selected)
    )
    pass

# функция принимает на вход модель и подготовленный для нее набор признаков
def plot_shap_bar(model, X_transformed_selected):
    # используем библиотеку shap
    explainer = shap.LinearExplainer(model, X_transformed_selected)
    shap_values = explainer(X_transformed_selected)
    shap.plots.bar(
        shap_values,
        # передаем размерность в max_display
        max_display=len(X_train_transformed_selected)
    )
    pass
```

```python
plot_shap_beeswarm(
    # передаем лучшую модель
    model=pipeline['classifier'],
    X_transformed_selected=X_train_transformed_selected
)
```

Построим столбчатая диаграмма `shap.plots.bar`, которая отражает общий вклад признаков в прогнозы модели. Она показывает средние SHAP-значения признака по всем наблюдениям.

### Краткий обзор проведенной работы:

Главные задачи проекта это построение модели машинного обучения, которая предскажет вероятность снижения покупательской активности клиента, и разработка на основе данных моделирования и прибыльности клиентов персонализированных предложений для повышения покупательской активности для определенного сегмента покупателей.

**В качестве исходных данных имеется 4 датасета:**
- таблица, которая содержит данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении
- таблица с данными о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом
- таблица с данными о времени (в минутах), которое покупатель провёл на сайте в течение периода
- таблица с данными о среднемесячной прибыли покупателя за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю


**Проведена предобработка данных:**
- устранена проблема с неявными дубликатами
- нейминг столбцов приведен в соответствие принятым соглашениям (`camel_case`)
- типы данных приведены в соответствие
- устранены опечатки в значениях

Для поиска лучшей модели был применен инструмент `GridSearchCV` и перебор гиперпараметров для четырех моделей:
- `DecisionTreeClassifier()`
- `KNeighborsClassifier()`
- `LogisticRegression()`
- `SVC()`

Лучшая модель отбиралась по метрике **RECALL**. Лучшей моделью оказалась `LogisticRegression()` с `L1-регуляризацией`. Для отбора признаков был использован инструмент `SelectKBest`.

- Метрика RECALL на кросс-валидации: **0.8**
- Метрика RECALL на тестовой выборке: **0.83**
- Снижение предела отнесения наблюдения к классу 1 несет большие издержки в общей точности модели.
